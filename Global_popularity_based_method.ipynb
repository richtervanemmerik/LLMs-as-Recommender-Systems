```python
import json
import pandas as pd
```

```python
df = pd.read_csv('data_user_items.csv')
df_labels = pd.read_csv('data_last_item_user_bought.csv')  
df
```

```python
popularity_df = df.groupby('title').size().reset_index(name='popularity_score')
popularity_df
```

```python
popularity_df.sort_values(by='popularity_score', ascending=False)
```

```python
def recommend_items(popularity_df, top_n):
    popularity_rank = popularity_df.sort_values(by='popularity_score', ascending=False)
    return popularity_rank.head(top_n)['title'].values
recommended_items = recommend_items(popularity_df, 10)  
global_recommended_items = ", ".join(recommended_items)

```

```python
global_recommended_items
```

```python
sample_df = df_labels.sample(n=150, random_state=42)
```

```python
correct = 0
for title in sample_df['title']:
    if title in global_recommended_items:
        correct += 1
print("Amount correct: ", correct)
accuracy = correct / len(df_labels)
print("Accuracy:", accuracy)        
```
