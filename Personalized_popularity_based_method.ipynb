```python
import json
import pandas as pd
```

```python
df = pd.read_csv('data_user_items.csv')
df_labels = pd.read_csv('data_last_item_user_bought.csv')  
```

```python
user_id = df['reviewerID'][4524]
```

```python
popularity_df = df.groupby(['reviewerID', 'title']).size().reset_index(name='popularity_score')
popularity_df
```

```python
sample_labels = df_labels.sample(n=150, random_state=42)
sample_labels
```

```python
def recommend_items(user_id, top_n):
    user_products = df[df['reviewerID'] == user_id]['title']
    user_unseen_products = df[~df['title'].isin(user_products)]['title']
    user_popularity = popularity_df[(popularity_df['title'].isin(user_unseen_products))]
    user_popularity = user_popularity.sort_values(by='popularity_score', ascending=False)
    return user_popularity.head(top_n)['title'].values
recommended_items = recommend_items(user_id, 10)  
personlized_recommended_items = ", ".join(recommended_items)    
print(personlized_recommended_items)
```

```python
correct = 0
for index, (user_id, title) in enumerate(sample_labels[['reviewerID', 'title']].values):
    if index % 10 == 0:
        print(f"{index} items for users predicited")    
    recommended_items = recommend_items(user_id, 10)
    personlized_recommended_items = ", ".join(recommended_items)
    if title in personlized_recommended_items:
        correct += 1
print("Amount correct: ", correct)
accuracy = correct / len(df_labels)
print("Accuracy:", accuracy)    
```
